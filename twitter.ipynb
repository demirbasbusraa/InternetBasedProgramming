## Takipçi Segmentasyonu

**Seçilen kullanıcıyı takip eden kullanıcıların bulunduğu dataframe oluşturmuştuk.**

```python
df.head()
```

**Kullanım kolaylığı açısından index olarak index sayısı değil screen_name bilgisini kullanıcağız.**

```python
df.index = df["screen_name"]
```

```python
df.head()
```

**Segmentasyon bilgisi için oluşturacağımız skor takipçi sayısı ve atılan tweet sayısının çarpımından oluşucaktır.**

```python
s_data = df[["followers_count", "statuses_count"]]
```

**Dataframe mı incelediğimizde sayılarla ilgili bir problemle karşılaşıyoruz. Skor takipçi sayısı ve tweet sayısının çarpımından oluşacaktır. Fakat bu sayılar içinde 0 bulunması bizim doğru sonuç almamızı engelliyor.**

```python
s_data
```

**Bu problemi çözmek için bütün takipçi ve tweet sayısı satırlarına 0.01 değeri ekledik. Bu çözüm sıfırın çarpmadaki etkisini yok etmiş ve sonucu çok değiştirmemiştir.**

```python
s_data["followers_count"] = s_data["followers_count"] + 0.01
s_data["statuses_count"] = s_data["statuses_count"] + 0.01
```

```python
s_data
```

**Örnek skor değerlerini inceledik.**

```python
6.01 * 1.01 
```

```python
4.01 * 39.01
```

**Örnek skorları incelediğimizde çıkan değerlerin segmentasyon yapılabilmesi için standartlaştırılması gerektiğini fark ettik.**

**Bunun için herbir değer üzerinde  (x-min(x))/(max(x)-min(x))) işlemini uygulayarak her değerin 0 ile 1 arasında bir sayı gelmesini sağladık.**

```python
s_data = s_data.apply(lambda x: (x-min(x))/(max(x)-min(x)))
```

```python
s_data
```

**Sonucu incelediğimizde tekrar 0 probleminin geldiğini gördük. Bunu çözmek için yine her değere 0.01 ekledik.**

```python
s_data["followers_count"] = s_data["followers_count"] + 0.01
s_data["statuses_count"] = s_data["statuses_count"] + 0.01
```

```python
s_data.head()
```

**Skor değerini takipçi ve tweet sayısı çarpımından oluşturduk.**

```python
skor = s_data["followers_count"] * s_data["statuses_count"]
```

```python
skor
```

**Skorları azalan şekilde inceledik.**

```python
skor.sort_values(ascending = False)
```

**Segmente ayırmak için medyanı kesme noktası olarak kullanıyoruz.**

```python
skor[skor > skor.median() + skor.std()/len(skor)].sort_values(ascending = False)
```

**Tabloya skor bilgisini de ekledil.**

```python
s_data["skor"] = skor
```

```python
s_data
```

**"s_data["skor"] >= skor.median() + skor.std()/len(skor)" koşuluna göre segmentasyon yapılır. Koşula uyanlar A, uymayanlar B ile işaretlenir.**

```python
import numpy as np
s_data["segment"] = np.where(s_data["skor"] >= skor.median() + 
                             skor.std()/len(skor), "A","B")
```

# Twitter Sentiment Analizi

**#datascience hashtag inden çekilen dataframe**

```python
df
```

```python
from textblob import TextBlob
```

```python
def sentiment_skorla(df):

    text = df["text"]

    for i in range(0,len(text)):
        textB = TextBlob(text[i])
        sentiment_skoru = textB.sentiment.polarity
        df.set_value(i, 'sentiment_skoru', sentiment_skoru)
        
        if sentiment_skoru <0.00:
            duygu_sinifi = 'Negatif'
            df.set_value(i, 'duygu_sinifi', duygu_sinifi )

        elif sentiment_skoru >0.00:
            duygu_sinifi = 'Pozitif'
            df.set_value(i, 'duygu_sinifi', duygu_sinifi )

        else:
            duygu_sinifi = 'Notr'
            df.set_value(i, 'duygu_sinifi', duygu_sinifi )
            
    return df 
```
